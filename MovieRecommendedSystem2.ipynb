```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
df = pd.read_csv("movie_dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.fillna(" ")
```

```python
df.columns
```

Features

```python
features = ['genres','keywords','cast','director']#drop overview
```

```python
def combine(row):
  return row['genres']+" "+row['keywords']+" "+row['cast']+" "+row['director']
```

```python
df['combine_features'] = combine(df[features])
```

```python
df.columns
```

Vectorizing and find the cosine similarity

```python
cv = CountVectorizer()
count = cv.fit_transform(df['combine_features'])
```

```python
similar = cosine_similarity(count)
```

```python
def get_index(title):
  return df[df.title == title].index.values[0]
def get_title(index):
  return df[df.index == index]["title"].values[0]
```

```python
movie_index = get_index("Pirates of the Caribbean: At World's End")
movie_index
```

```python
similar_movies = list(enumerate(similar[movie_index]))
```

**Sorting**




```python
sorted_movie = sorted(similar_movies,key=lambda x:x[1],reverse=True)
sorted_movie
```

```python
for i in range(20):
  print(get_title(sorted_movie[i][0]))
```
